# Task 1: Exploring and Visualizing a Simple Dataset (Iris Dataset)
## Objective: Load, inspect, and visualize the Iris dataset to understand data trends and distributions.


Step 1: Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Enabling inline plotting
%matplotlib inline

# seaborn style for better visuals
sns.set(style="whitegrid")
```

Step 2: Load the Dataset

```python
# Load the built-in Iris dataset from seaborn
df = sns.load_dataset("iris")

# Display the first 5 rows
df.head()
```

Step 3: Inspect the Dataset

```python
# Shape of the dataset (rows, columns)
print("Shape of dataset:", df.shape)

# Column names
print("\nColumn names:", df.columns.tolist())

# Data types and non-null counts
print("\nDataset Info:")
print(df.info())
```

Step 4: Summary Statistics

```python
# Summary statistics for numerical features
df.describe()
```

Step 5: Data Visualization

```python
# Scatter plot: sepal_length vs sepal_width, color by species
plt.figure(figsize=(8,6))
sns.scatterplot(x="sepal_length", y="sepal_width", hue="species", data=df, s=70)
plt.title("Sepal Length vs Sepal Width by Species")
plt.show()
```

```python
# Histograms for all numeric columns
df.hist(figsize=(10,8), color='skyblue', edgecolor='black')
plt.suptitle("Feature Distributions")
plt.show()
```

```python
# Box plots for each feature grouped by species
plt.figure(figsize=(10,6))
sns.boxplot(data=df)
plt.title("Boxplot of All Numeric Features")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="species", y="sepal_length", data=df)
plt.title("Sepal Length by Species")
plt.show()
```

Step 6: Observations / Insights

### Insights:
- The dataset has 150 entries and 5 columns (4 numeric features + 1 categorical).
- Each Iris species shows distinct patterns in petal and sepal measurements.
- Histograms show normally distributed features with slight variation.
- Boxplots indicate minimal outliers, mostly in sepal width.
- Visual separation suggests petal length and petal width are good discriminators of species.
